```python
from pytesseract import Output
from PIL import Image

import matplotlib.pyplot as plt
import pytesseract
import cv2
import numpy as np
import pandas as pd
import csv
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
filename = 'wwwroot/assets/nb_example_01.jpeg'
img1 = np.array(Image.open(filename))
text = pytesseract.image_to_string(img1)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
print(text)
```

```python
filename = 'wwwroot/assets/nb_example_01_noised.png'
img2 = np.array(Image.open(filename))
text = pytesseract.image_to_string(img2)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
print(text)
```

```python
norm_img = np.zeros((img2.shape[0], img2.shape[1]))
img2 = cv2.normalize(img2, norm_img, 0, 255, cv2.NORM_MINMAX)
img2 = cv2.threshold(img2, 100, 255, cv2.THRESH_BINARY)[1]
img2 = cv2.GaussianBlur(img2, (1, 1), 0)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
text = pytesseract.image_to_string(img2)
print(text)
```
